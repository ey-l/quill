```python
songs_df = pd.read_csv("./songs/input/songs_raw.csv")
# print columns
print(songs_df.columns)
```

```python
songs_df.describe()
```

```python
# Make the table songs_df.describe() into a string
songs_description = songs_df.describe().to_string()
songs_description
```

```python
songs_df.head()
```

```python
# bin song_popularity into 6 bins, based on the quantiles 25%, 50%, 75%, 90%, 95%
songs_df['popularity_percentile'] = pd.qcut(songs_df['song_popularity'],
                                     q=[0, 0.25, 0.5, 0.75, 0.9, 0.95, 1],
                                     labels=['very_low', 'low', 'medium', 'high', 'very_high', 'top'])
# save the songs_df with the new column to a new csv file
songs_df.to_csv("./songs/input/songs.csv", index=False)
```

```python
taxi_df = pd.read_csv("./taxi/input/taxi_raw.csv")
# print columns
print(taxi_df.columns)
```

```python
taxi_df.describe()
```

```python
taxi_df.describe().to_string()
```

```python
taxi_df["fare_percentile"] = pd.qcut(taxi_df["fare_amount"],
                                     q=[0, 0.25, 0.5, 0.75, 0.9, 0.95, 1],
                                     labels=['very_low', 'low', 'medium', 'high', 'very_high', 'top'])
# Get dummies for the payment_type column
taxi_df = pd.get_dummies(taxi_df, columns=['payment_type'], prefix='payment_type')
print(taxi_df.columns)
# save the taxi_df with the new column to a new csv file
taxi_df.to_csv("./taxi/input/taxi.csv", index=False)
```

```python
bank_df = pd.read_csv("./bank/input/bank_raw.csv")
# print columns
print(bank_df.columns)
```

```python
bank_df.describe()
```

```python
bank_df.describe().to_string()
```

```python
bank_df.dropna(subset=['Loan Status'], inplace=True)
bank_df.drop_duplicates(inplace=True)
```

```python
missing_values_count = bank_df.isnull().sum()
print("Count of missing values in each column:")
print(missing_values_count)
```

```python
# Drop nulls in ["Credit Score", "Annual Income", "Years in current job", "Bankruptcies", "Tax Liens"]
bank_df.dropna(subset=["Credit Score", "Annual Income", "Years in current job", "Maximum Open Credit", "Bankruptcies", "Tax Liens"], inplace=True)
missing_values_count = bank_df.isnull().sum()
print("Count of missing values in each column:")
print(missing_values_count)
print(len(bank_df))
```

```python
# Fill nulls in "Credit Inquiry" with 0
bank_df["Months since last delinquent"].fillna(0, inplace=True)
```

```python
bank_df.head()
```

```python
# Unique  values in "Loan Status"
print(bank_df["Loan Status"].unique())
```

```python
# Change "Loan Status" to 0 and 1
bank_df["Loan Status"] = bank_df["Loan Status"].map({"Fully Paid": 0, "Charged Off": 1})
```

```python
bank_df.describe()
```

```python
# Save the bank_df to a new csv file
bank_df.to_csv("./bank/input/bank.csv", index=False)
```
